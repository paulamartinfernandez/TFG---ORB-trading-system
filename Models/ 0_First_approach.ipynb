BP (British Pound) tanto min como dia

```python
df_bp1min = pd.read_csv(path+'/bp1min.txt')
df_bp1min
```

```python
df_bp1min['datetime'] = pd.to_datetime(
    df_bp1min['<DTYYYYMMDD>'].astype(str) + df_bp1min['<TIME>'].astype(str).str.zfill(6),
    format='%Y%m%d%H%M%S'
)
df_bp1min = df_bp1min.sort_values('datetime')
df_bp1min = df_bp1min.drop(columns=['<DTYYYYMMDD>'])
df_bp1min = df_bp1min.dropna()
```

```python
df_bp1min
```
